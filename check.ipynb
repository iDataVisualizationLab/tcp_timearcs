```python
import pandas as pd

df90 = pd.read_csv('90min_day1.csv')

df90.head()
```

```python
first_ts = df90['timestamp'].iloc[0]
print(first_ts)
df90 = df90.iloc[1:].copy()
df90['timestamp'] = first_ts + df90['timestamp']
df90.head()

```

```python
# Remove rows with attack == -1 (or matched_event == -1) before deduplication
df90_min = df90.copy()
col_to_check = 'attack' if 'attack' in df90_min.columns else ('matched_event' if 'matched_event' in df90_min.columns else None)
if col_to_check is None:
    raise KeyError("No 'attack' or 'matched_event' column found to filter -1 values.")
before_rows = len(df90_min)
df90_min = df90_min[df90_min[col_to_check] != -1].copy()
removed = before_rows - len(df90_min)
print(f"Removed {removed} rows where {col_to_check} == -1 (from {before_rows} -> {len(df90_min)})")

# represent timestamp as minutes (no seconds / milliseconds)
df90_min['timestamp'] = df90_min['timestamp'] // 60000

# Deduplicate on selected columns (ensure dupe_cols excludes 'count')
dupe_cols = ['timestamp', 'length', 'src_ip', 'dst_ip', 'protocol', 'attack']

# Build stable key and aggregate
key_col = '__key__'
df90_min[key_col] = pd.util.hash_pandas_object(df90_min[dupe_cols], index=False)
first_mask = ~df90_min[key_col].duplicated()
count_sum = df90_min.groupby(key_col, sort=False)['count'].sum()

df90_minute_combined = df90_min.loc[first_mask, dupe_cols].copy()
df90_minute_combined['count'] = df90_min.loc[first_mask, key_col].map(count_sum).values
df90_minute_combined.reset_index(drop=True, inplace=True)

print(f'Rows after combine: {len(df90_minute_combined)}  Collapsed: {len(df90_min) - len(df90_minute_combined)}')

df90_minute_combined.to_csv('90min_day1_collapsed.csv', index=False)
df90_minute_combined.head()
```
